# All of the text blocks used by filers

```python
import calcbench as cb
```

```python
companies = cb.companies(index='DJIA')
```

```python
text_blocks = {}
for company in companies.to_dict('records'):
    docs = cb.document_search(company_identifiers=[company['entity_id']], 
                              year=2017, 
                              period=0, 
                              all_documents=True,
                              period_type='Annual')
    text_blocks[company['cik_code']] = [d['local_name'] for d in docs if d['local_name']]
```
